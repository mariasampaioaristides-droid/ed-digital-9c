## Jogo da Memória Numérica (Versão 2.0)

Este jogo consiste em memorizar uma sequência numérica que cresce a cada acerto. A primeira sequência terá 5 números. O tempo para inserir a sequência também aumenta gradualmente. Se você acertar 10 rodadas, você ganha! Se errar, o jogo reinicia.

```python
import random
import time

def play_memory_game():
    while True: # Loop externo para permitir reiniciar o jogo
        score = 0
        sequence_length = 5  # Requisito: A primeira sequência terá 5 números
        time_limit = 5  # segundos
        game_on = True
        max_rounds = 10 # Requisito: Máximo de 10 acertos

        print("\n--- Novo Jogo ---")
        print("Bem-vindo ao Jogo da Memória Numérica!")
        print("Lembre-se da sequência de números e digite-a corretamente.")
        print(f"A cada acerto, a sequência fica mais longa e você terá mais 1 segundo. Máximo de {max_rounds} acertos.")

        while game_on:
            if score >= max_rounds:
                print("\nParabéns! Você ganhou! Você acertou 10 rodadas!")
                game_on = False
                continue # Sai do loop interno do jogo para perguntar se quer jogar de novo

            print(f"\n--- Nível {score + 1} ---")
            print(f"Comprimento da sequência: {sequence_length} números")
            print(f"Tempo limite para resposta: {time_limit} segundos")

            # Gerar sequência
            sequence = [random.randint(0, 9) for _ in range(sequence_length)]
            sequence_str = " ".join(map(str, sequence))

            print("Memorize esta sequência:")
            print(sequence_str)

            # Tempo de exibição da sequência (ajustado para ser um pouco mais longo para sequências maiores)
            display_time = max(2, min(sequence_length * 0.75, 5)) # Min 2s, Max 5s, 0.75s por dígito
            time.sleep(display_time)

            # Simular limpeza da tela para esconder a sequência
            print("\n" * 50)
            print("Qual foi a sequência?")

            start_time = time.time()
            user_input = input("Sua resposta (números separados por espaço): ")
            end_time = time.time()

            time_taken = end_time - start_time

            # Processar entrada do usuário
            try:
                user_sequence = [int(x) for x in user_input.split()]
            except ValueError:
                user_sequence = [] # Entrada inválida

            if time_taken > time_limit:
                print(f"Tempo esgotado! Você levou {time_taken:.2f} segundos de {time_limit} permitidos.")
                game_on = False
            elif user_sequence == sequence:
                print("Correto!")
                score += 1
                sequence_length += 1
                time_limit += 1  # Aumenta o tempo limite em 1 segundo
            else:
                print("Incorreto! A sequência correta era:", sequence_str)
                game_on = False

        print(f"\n--- Fim de Jogo ---")
        print(f"Sua pontuação final: {score}")

        play_again = input("Deseja jogar novamente? (s/n): ").lower()
        if play_again != 's':
            break # Sai do loop externo se o usuário não quiser jogar de novo

# Para iniciar o jogo, remova o '#' antes da linha abaixo e execute a célula:

play_memory_game()
```
